```python
url = requests.get('https://realpython.com/atom.xml')
soup = BeautifulSoup(url.content, 'xml')
entries = soup.find_all('entry')
```

```python
for entry in entries:
    title = entry.title.text
    summary = entry.summary.text
    link = entry.link['href']
    print(f"\tTitle: {title}\n\n\tSUMMARY:\n {summary}\n\n\tLink: {link}\n\n\n\t\t\t\t\t\t..........................\n\n")
    
   
```

```python
import pandas as pd
```

```python
data = {}
for entry in entries:
    data.setdefault('title',[])
    data.setdefault('summary',[])
    data.setdefault('link',[])
    data['title'].append(entry.title)
    data['summary'].append(entry.summary)
    data['link'].append(entry.link)
    
```

```python
df = pd.DataFrame(data)
```
